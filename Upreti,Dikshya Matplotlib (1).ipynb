```python
data = df['gross income']
plt.hist(data, bins = np.arange(min(data), max(data) + 1, 1))
plt.title("Histogram Showing Gross Income For Supermarket Sales")
plt.hist(data, density = True)
plt.ylabel('Density')
plt.xlabel('Gross_income')
```

```python
sns.countplot(y = 'Payment', data=df, order = df['Payment'].value_counts().index )
```

```python
sns.boxplot(x=df['Gender'], y=df['Total'])
plt.title('Boxplot Showing Genders')
sns.set_style('whitegrid')
```

```python
 sns.countplot(x="Customer type", hue = "Branch", data =df).set_title("Customer Type by Branch") 
```

```python
cols = ['Product line']
numerical = cols

plt.figure(figsize=(20,4))

for i, col in enumerate(numerical):
    ax = plt.subplot(1, len(numerical), i+1)
    sns.countplot(x=str(col), data=df)
    ax.set_title(f"{col}")
```

```python
plt.title('Boxplot of quantity sales by product line')
sns.boxenplot(y = 'Product line', x = 'Quantity', data=df )
```

```python
# Scatterplot arguments
sns.lmplot(x='Total', y='Unit price', data=df, fit_reg=False) 
plt.title('Scatter Plot Showing Unit And Total Price')
plt.show()
```

```python
plt.figure(figsize = (15, 12))
sns.heatmap(df.corr(), cmap="RdYlGn", annot=True, fmt=".1f")
plt.title('Heatmap Showing Numerical Variables')
plt.show()
```

```python
sns.relplot(data=df, x="City", y="Unit price", kind="line")
plt.title("Unit Price By City")

```

```python
 sns.swarmplot(x="Customer type",  y = "Rating",  hue = "City", data =df).set_title("Customer Ratings By City & Membership Type") 
```
